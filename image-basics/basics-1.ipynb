<h2>Threashold</h2>

```python
# 125 threhold value
# 255 max value
# threshold type
# any pixel value below 125 becomes 0 and any pixel above 125 becomes 255

import cv2

image_path='../images/bunny.jpeg'

#the zero at the end makes the image b&w
image=cv2.imread(image_path,0)


ret, thresh1 = cv2.threshold(image, 125,255, cv2.THRESH_BINARY)
cv2.imshow("thresh1",thresh1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 125 threhold value
# 255 max value
# threshold type
# any pixel value above 125 becomes 0 and any pixel below 125 becomes 255

ret, thresh2 = cv2.threshold(image, 100,255, cv2.THRESH_BINARY_INV)
cv2.imshow("thresh2",thresh2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

<h2>Masks</h2>

```python
import cv2
import numpy as np


image = cv2.imread('../images/bunny.jpeg')
# image.shape

# Create a black image of the same image you want to apply the mask to
mask = np.zeros(image.shape[:2], dtype=np.uint8)

# create mask by choosing range of white pixels
mask[320:920,180:540]=255

# Display the mask
cv2.imshow('Mask', mask)
while True:
    if cv2.waitKey(100) & 0xFF == ord('q'):  
        break
cv2.destroyAllWindows()
```

```python
# to apply mask to your image, bitwise operation and is used
# black and any color = black 
# white and any color = any color
masked_image= cv2.bitwise_and(image, image, mask=mask)
cv2.imshow('Image', image)
cv2.imshow('Mask', mask)
cv2.imshow('Masked Image', masked_image)
while True:
    if cv2.waitKey(100) & 0xFF == ord('q'):  
        break
cv2.destroyAllWindows()
```

<h2>Color Extraction</h2>

```python
import numpy as np

image = cv2.imread("../images/bunny.jpeg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

plt.imshow(image)
plt.show()

plt.imshow(hsv_image)
plt.show()
lower_bound = np.array([0,100,100])
upper_bound = np.array([180,255,255])

extracted_image = cv2.inRange(hsv_image, lower_bound, upper_bound)

masked_image= cv2.bitwise_and(image, image, mask=extracted_image)


plt.imshow(masked_image)
plt.show()


```

<h3>get points using mouse events</h3>

```python

import cv2
point1 = None
point2 = None

def getPoints(event, x, y, flags, param):
    global point1, point2
    if event ==cv2.EVENT_LBUTTONDOWN:
        if point1 is None:
            point1 = (x, y)
            print(f"point 1: {point1}")
        elif point2 is None:
            point2 = (x, y)
            print(f"point 2: {point2}")

image_path='../images/bunny.jpeg'
image=cv2.imread(image_path,0)
cv2.namedWindow('Image')
cv2.setMouseCallback('Image', getPoints)


while True:
    cv2.imshow('Image',image)
    
    key = cv2.waitKey(1)
    if key == 27:
        break
cv2.destroyAllWindows()
```
